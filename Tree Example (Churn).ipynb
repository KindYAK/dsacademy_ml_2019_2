```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("telecom_churn.csv")
```

```python
df_x = df.drop(['Churn', 'State'], axis=1)
df_y = df['Churn']
```

```python
def f(x):
    return 1 if x == "Yes" else 0
df_x['International plan'] = df_x['International plan'].apply(f)
df_x['Voice mail plan'] = df_x['Voice mail plan'].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
for depth in range(1, 50):
    classifier = DecisionTreeClassifier(max_depth=depth)
    classifier = classifier.fit(X_train, y_train)
    print(depth, classifier.score(X_train, y_train), classifier.score(X_test, y_test))
```
