```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("telecom_churn.csv")
```

```python
df.head()
```

```python
df['Customer service calls'].plot.hist()
```

```python
df_l = df[df['Customer service calls'] <= 2]
df_h = df[df['Customer service calls'] >= 3]
```

```python
print(df['Churn'].value_counts()[True]/len(df))
print(df_l['Churn'].value_counts()[True]/len(df_l))
print(df_h['Churn'].value_counts()[True]/len(df_h))
```

```python
df.groupby('Churn').count()['State'].plot.pie()
```

```python
df_churn = df[df['Churn']==True]
df_non_churn = df[df['Churn']==False]
```
